# histogram (normalizing sample size)

when we are analyzing two sets of data with diffeent sample size, it will be difficult to analyze the visualized data.
We can normalize the data samples using the function below:

plt.hist(data1, density=True) |or|  plt.hist(data1, density=1)
plt.hist(data2, density=True) |or|  plt.hist(data2, density=1)

```python
#visualizing data frame using matplotlib library
import matplotlib.pyplot as plt
dogs['height_cm'].hist()
plt.show()

#x-axis represents the heights of the dogs
#y-axis represents the number of dogs in each height range
#result shows that there are alot of dogs with heights between 40 and 60
```

```python
#the number of bars/bins can be adjusted using the bins argument (eg: bins=5 or 10)
dogs['height_cm'].hist(bins=5)
```

```python
#visualizing the avrage (mean) weight of dog breeds
#line graph
avg_weight_by_breed = dogs.groupby('breed')['weight_kg'].mean()
print(avg_weight_by_breed)
avg_weight_by_breed.plot() #without specifying the type of graph, the data will be visialized as a line graph.
plt.title('Average Weight by Dog Breed') #adding title to the graph
```

```python
#bar chart
#using the kind function in the plot bracket specifies what kind of graph you want to be displayed
#alternative of adding title for a bar plot
avg_weight_by_breed.plot(kind='bar', title='Average Weight by Dog Breed')
```

```python
#bar chart by selecting x and y from data frame
#bar graph of dog breeds by weight
dogs.plot(x='breed', y='weight_kg', kind='bar', title='Breed by Weight')
```
